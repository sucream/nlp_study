# 1. TF 만들기
- 각 문서를 형태소분석, stopword 처리 후 BOW로 만든다
- 각 단어들의 빈도를 파악하기 위해 Counter를 사용하여 빈도 체크
- 단어의 빈도수 / 전체 단어수 = TF

## 1.1. 형태소 분석 및 하나의 데이터로 만들기

```python
import pos_tag
```

```python
korea1_pos = pos_tag.new_pos(korea1_text.splitlines(), discard_stopwords=True, discard_verb=True)
korea1_all_doc_tokens = sum(korea1_pos, [])
korea1_all_doc_tokens
```

```python
korea2_pos = pos_tag.new_pos(korea2_text.splitlines(), discard_stopwords=True, discard_verb=True)
korea2_all_doc_tokens = sum(korea2_pos, [])
korea2_all_doc_tokens
```

```python
chosun_pos = pos_tag.new_pos(chosun_text.splitlines(), discard_stopwords=True, discard_verb=True)
chosun_all_doc_tokens = sum(chosun_pos, [])
chosun_all_doc_tokens
```

## 1.2. Counter를 이용한 빈도 확인

```python
from collections import Counter
```

```python
korea1_bow = Counter(korea1_all_doc_tokens)
korea2_bow = Counter(korea2_all_doc_tokens)
chosun_bow = Counter(chosun_all_doc_tokens)
```

```python
korea1_bow.most_common(5)
```

```python
korea2_bow.most_common(5)
```

```python
chosun_bow.most_common(5)
```

### lexicon 만들기

```python
import pandas as pd
```

```python
lexicon = set()
lexicon.update(korea1_all_doc_tokens)
lexicon.update(korea2_all_doc_tokens)
lexicon.update(chosun_all_doc_tokens)
lexicon
```

## 1.3. TF 값 구하기

```python
korea1_tf = {}
for i in korea1_bow:
    korea1_tf[i] = korea1_bow[i]/len(korea1_bow)
#     print(i, korea1_bow[i]/len(korea1_bow))
pd.DataFrame(korea1_tf, index=['korea1_tf']).T
```

```python
korea2_tf = {}
for i in korea2_bow:
    korea2_tf[i] = korea2_bow[i]/len(korea2_bow)
#     print(i, korea2_bow[i]/len(korea2_bow))
pd.DataFrame(korea2_tf, index=['korea2_tf']).T
```

```python
chosun_tf = {}
for i in chosun_bow:
    chosun_tf[i] = chosun_bow[i]/len(chosun_bow)
#     print(i, chosun_bow[i]/len(chosun_bow))
pd.DataFrame(chosun_tf, index=['chosun_tf']).T
```

## 2. IDF 만들기
- log(전체 문서의 수 / 해당 단어를 포함한 문서의 수) = IDF

```python
import math
num_of_all_documents = 3
```

```python
korea1_idf = {}

for i in korea1_all_doc_tokens:
    num_of_count = 0
    if i in korea1_all_doc_tokens:
        num_of_count += 1
    if i in korea2_all_doc_tokens:
        num_of_count += 1
    if i in chosun_all_doc_tokens:
        num_of_count += 1
    
    korea1_idf[i] = math.log10(num_of_all_documents / num_of_count)
    
pd.DataFrame(korea1_idf, index=['korea1_idf']).T
```

```python
korea2_idf = {}

for i in korea2_all_doc_tokens:
    num_of_count = 0
    if i in korea1_all_doc_tokens:
        num_of_count += 1
    if i in korea2_all_doc_tokens:
        num_of_count += 1
    if i in chosun_all_doc_tokens:
        num_of_count += 1
    
    korea2_idf[i] = math.log10(num_of_all_documents / num_of_count)
    
pd.DataFrame(korea2_idf, index=['korea2_idf']).T
```

```python
chosun_idf = {}

for i in chosun_all_doc_tokens:
    num_of_count = 0
    if i in korea1_all_doc_tokens:
        num_of_count += 1
    if i in korea2_all_doc_tokens:
        num_of_count += 1
    if i in chosun_all_doc_tokens:
        num_of_count += 1
    
    chosun_idf[i] = math.log10(num_of_all_documents / num_of_count)
    
pd.DataFrame(chosun_idf, index=['chosun_idf']).T
```

# 3. TF-IDF 구하기
- tf와 idf의 값을 곱한 값 = tf-idf

```python
korea1_tfidf = {}

for i in korea1_all_doc_tokens:
    korea1_tfidf[i] = korea1_tf[i] * korea1_idf[i]
    
pd.DataFrame(korea1_tfidf, index=['korea1_tfidf']).T
```

```python
korea2_tfidf = {}

for i in korea2_all_doc_tokens:
    korea2_tfidf[i] = korea2_tf[i] * korea2_idf[i]
    
pd.DataFrame(korea2_tfidf, index=['korea2_tfidf']).T
```

```python
chosun_tfidf = {}

for i in chosun_all_doc_tokens:
    chosun_tfidf[i] = chosun_tf[i] * chosun_idf[i]
    
pd.DataFrame(chosun_tfidf, index=['chosun_tfidf']).T
```

```python
all_tfidf = [{i: 0 for i in lexicon}, {i: 0 for i in lexicon}, {i: 0 for i in lexicon}]

for i in korea1_tfidf:
    all_tfidf[0][i] = korea1_tfidf[i]
for i in korea2_tfidf:
    all_tfidf[1][i] = korea2_tfidf[i]
for i in chosun_tfidf:
    all_tfidf[2][i] = chosun_tfidf[i]

data = pd.DataFrame(all_tfidf, index=['korea1', 'korea2', 'chosun']).T
```

## TF-IDF 값을 기준으로 코사인 유사도 확인

```python
import math
import numpy as np

def cosine_sim(vec1, vec2):
    dot_prod = 0
    
    for i, v in enumerate(vec1):
        dot_prod += v * vec2[i]  # 내적
    
    mag_1 = np.linalg.norm(vec1)
    mag_2 = np.linalg.norm(vec2)
    
    return dot_prod / (mag_1 * mag_2)  # 코사인세타 공식
```
